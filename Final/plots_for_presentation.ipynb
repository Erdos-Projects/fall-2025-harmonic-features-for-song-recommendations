```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#import raw data
path_to_raw_data='../../data/chordonomicon_raw.csv'
path_to_final_training_data='../../data/data_train_all.csv'
raw_df=pd.read_csv(path_to_raw_data, low_memory=False)
final_train_df=pd.read_csv(path_to_final_training_data, low_memory=False)
```

```python
print(raw_df.shape)
print(final_train_df.shape)
```

```python
yrs=final_train_df['decade'].value_counts()
plt.bar(yrs.index,yrs,width=4)
plt.xlabel('Decade')
plt.ylabel('Number of Songs')
plt.show()
```

```python
grs=final_train_df['main_genre'].value_counts()
plt.barh(grs.index,grs)
plt.xlabel('Number of Songs')
plt.ylabel('Main Genre')
```

```python
print(final_train_df.columns[:30])
```

```python
plt.hist(final_train_df['spotify_popularity'],bins=30)
plt.xlabel('Spotify popularity score')
plt.ylabel('Total number of songs')
plt.show()
```
