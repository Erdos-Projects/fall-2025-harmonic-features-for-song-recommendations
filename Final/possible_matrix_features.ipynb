This notebook creates possible features dervied from chord matrices.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import ast
```

```python
#import cleaned training dataset 
path_to_clean_df='../../data/final_train.csv'
df=pd.read_csv(path_to_clean_df)
```

Some final features to test:
* missing_notes: the number of notes not contained in any chord of the song (integer between 0 and 12)
* drone_ratio: ratio of the note used most often in chords to the total number of chords
* average_overlap: for each chord vector in the song, take the dot product with the following chord vector and average over all chord vectors
* average_noverlap, n=2,3,4,5: identical to average overlap, except that the dot product is taken between the $i^{th}$ and $i+n^{th}$ vector instead of the $i^{th}$ and $i+1^{th}$ vector for each $i$

Define some general helper functions:

```python
#helper function to modify the chords feature to help in analysis

# Read the mapping CSV file
chord_relations = pd.read_csv('../data/chords_mapping.csv')

# Create a dictionary with keys the "chords" and values the "degrees"
chord_degrees = dict(zip(chord_relations['Chords'], chord_relations['Degrees']))
for key, value in chord_degrees.items():
    chord_degrees[key] = ast.literal_eval(value)

# replacing spaces with commas
def replace_space_with_comma(my_string):
    return my_string.replace(" ",",")

# replacing spaces with commons in all chords in all rows of the data
df.loc[:,'chords'] = df['chords'].apply(replace_space_with_comma)

# Remove section markers
def remove_section_markers(my_string):
    result = []
    i = 0
    n = len(my_string)
    while i < n:
        if my_string[i] == '<':
            # Skip until after the following ", "
            j = my_string.find('>', i)
            if j == -1:
                break  # no closing '>', stop
            i = j + 2  # skip '>,' and the space
        else:
            result.append(my_string[i])
            i += 1
    assert('<' not in result)
    assert('>' not in result)
    return ''.join(result)

df.loc[:,'chords'] = df['chords'].apply(remove_section_markers)

# Removing inversions
def remove_inversions(my_string):
    result = []
    i = 0
    n = len(my_string)
    while i < n:
        if my_string[i] == '/':
            # Skip until after the following ", "
            j = my_string.find(',', i)
            if j == -1:
                break  # no closing comma, stop
            i = j  # skip comma
        else:
            result.append(my_string[i])
            i += 1
    return ''.join(result)

df.loc[:,'chords'] = df['chords'].apply(remove_inversions)
```

```python
# function to convert a string of comma-separated chords into a matrix, where each column denotes a chord

def string_to_chord_matrix(chord_sequence):
    # split sequence over commas, ignoring any "empty string" chords
    chord_list = [c for c in chord_sequence.split(',') if c != '']
    
    # then look up each chord in chord_degrees dictionary by the key value
    return np.array([chord_degrees[c][::-1] for c in chord_list]).transpose()

def total_chords(chord_string):
    # count the number of total chords in a comma-separated string
    return len(chord_string.split(','))

#create temporary chord_matrix and total_chords features
df.insert(loc = 1,
                  column = 'chord_matrix',
                  value = df['chords'].apply(string_to_chord_matrix),
                  allow_duplicates = False)
df.insert(2,'tot_chords',df['chords'].apply(total_chords))
```

Insert above features:

```python
#define functions to apply to chord_matrix feature

def missing_notes(chord_matrix):
    # returns the number of notes missing, i.e. number of zero rows in the chord matrix
    count = 0
    for row in chord_matrix:
        if not row.any():
            count += 1
    return count

def max(chord_matrix):
    # return a vector of note counts
    count_vector = np.array([np.sum(chord_matrix[i]) for i in range(12)])
    return np.max(count_vector)

def gen_overlap(chord_matrix,n):
    # a measure of how much sequential pairs of chords overlap in notes
    l=len(chord_matrix.T)
    similarity_vector = np.zeros(shape = l-n, 
                                 dtype=int)
    for i in range(l-n):
        # look the ith row (chord), and compare with the following chord
        # take the dot product, which counts the number of overlapping notes
        similarity_vector[i] = np.dot(chord_matrix[:,i],chord_matrix[:,i+n])
    print(similarity_vector)
    return np.mean(similarity_vector)


```

```python
#insert new features
df.insert(3,'missing_notes',df['chord_matrix'].apply(missing_notes))
df.insert(6,'drone_ratio',df['chord_matrix'].apply(max)/df['tot_chords'])
df.insert(7,'average_overlap',df['chord_matrix'].apply(gen_overlap,n=1))
df.insert(9,'average_2overlap',df['chord_matrix'].apply(gen_overlap,n=2))
df.insert(10,'average_3overlap',df['chord_matrix'].apply(gen_overlap,n=3))
df.insert(11,'average_4overlap',df['chord_matrix'].apply(gen_overlap,n=4))
df.insert(12,'average_5overlap',df['chord_matrix'].apply(gen_overlap,n=5))
```

Insert two additional features: 
* maj_triad_ratio: ratio of major triads to total number of chords per song
* min_triad_ratio: ratio of minor triads to total number of chords per song

```python
#define helper functions

from collections import deque 

def transpose_chord_up(chord_vector, num_semitones):
    # transpose the input chord_vector up by num_semitones
    d = deque(chord_vector)
    d.rotate(num_semitones)
    return(list(d))

def chord_ratio(mat,trans):
    #given a set of twelve transpositions of a chord, return the ratio 
    #of chords belonging to that set to the total number of chords
    chords=[list(row) in trans for row in np.transpose(mat)]
    return np.mean(chords)

```

```python
#create variables

#define sets of transpositions
maj_triad_vec=[0,0,1,0,0,1,0,0,0,1,0,0] #D
maj_transpositions=[]
for i in range(12):
    maj_transpositions.append(transpose_chord_up(maj_triad_vec,i))

min_triad_vec=[0,0,0,0,1,0,0,0,1,0,0,1] #Cmin
min_transpositions=[]
for i in range(12):
    min_transpositions.append(transpose_chord_up(min_triad_vec,i))
```

```python
#insert major triad ratio into df
df.insert(13,'maj_triad_ratio',df['chord_matrix'].apply(chord_ratio,trans=maj_transpositions))
df.insert(14,'min_triad_ratio',df['chord_matrix'].apply(chord_ratio,trans=min_transpositions))
```
