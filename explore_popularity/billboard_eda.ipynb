## Label songs by decade 

```python
df['chart_week'] = pd.to_datetime(df['chart_week'])
df['year'] = df['chart_week'].dt.year
df['month'] = df['chart_week'].dt.month
df['day'] = df['chart_week'].dt.day
df 
```

```python
df['decade'] = (df['year'] // 10 * 10).astype(str) + 's'
df.sample(10)
```

## Get a list of songs and artists that hit a peak position for a given decade

```python
def decade_peaks(pos, decade):
    df_filt = df[(df['peak_pos']<=pos) & (df['decade'] == decade)]
    song_list = sorted(df_filt['title'].unique())
    num_songs = len(df_filt['title'].unique())
    artist_list = sorted(df_filt['performer'].unique())
    num_artists = len(df_filt['performer'].unique())
    
    return song_list, num_songs, artist_list, num_artists

decade_peaks(1, '1990s')
```

### Or, a list of songs that debuted at #1 (assuming that NA indicates the song was not on the chart the previous week)



```python
def debuted_at_pos(pos, decade):
    df_filt = df[(df['peak_pos']<=pos) & (df['decade'] == decade) & (df['last_week'].isna())]
    song_list = sorted(df_filt['title'].unique())
    num_songs = len(df_filt['title'].unique())
    artist_list = sorted(df_filt['performer'].unique())
    num_artists = len(df_filt['performer'].unique())
    
    return song_list, num_songs, artist_list, num_artists

debuted_at_pos(1, '1990s')
```

## Decade Trends

```python
artists_per_decade = df.groupby('decade')['title'].nunique()
artists_per_decade.plot(kind="bar")
plt.ylabel("Unique Songs on Chart")
plt.title("Unique songs per decade")
```

### We can also look at unique #1 songs per decade 

```python
df_peak1 = df[df['peak_pos'] == 1]
num1_per_decade = df_peak1.groupby('decade')['title'].nunique()
num1_per_decade.plot(kind="bar")
plt.ylabel("Unique #1 Songs on Chart")
plt.title("Unique #1 songs per decade")

```

### As well as unique artists

```python
artists_per_decade = df.groupby('decade')['performer'].nunique()
artists_per_decade.plot(kind="bar")
plt.ylabel("Unique artists on chart")
plt.title("Unique artists per decade")
```

### As we saw from above, this could be due to features inflating the artist count. Let's see if songs featuring other artists changes by decade. This is not a perfect method, as some songs use & to denote that 2 artists worked on a song. Filtering out & is complicated because some band names have & in them 

```python
df_feat = df[df['performer'].str.contains('featur', case=False)]

features = df_feat.groupby('decade')['title'].nunique()
features.plot(kind="bar")
plt.ylabel("# of songs with features")
plt.title("Songs with features per decade")
```

```python
song_weeks = df.groupby(['title', 'performer'], as_index=False)['wks_on_chart'].max()
song_weeks = song_weeks.merge(df[['title', 'performer', 'decade']], on=['title', 'performer'], how='left').drop_duplicates(['title', 'performer'])

weeks_per_decade = song_weeks.groupby('decade')['wks_on_chart'].mean()

weeks_per_decade.plot(kind="bar", figsize=(8,5))
plt.ylabel("Mean weeks on chart")
plt.title("Chart longevity per decade")
plt.show()
```

### Finally, we can also look at whether songs tend to have wide or narrow range of peak chart positions by decade

```python
song_range = (
    df.groupby(['title', 'performer'], as_index=False)['peak_pos']
      .agg(lambda x: x.max() - x.min())
      .rename(columns={'peak_pos': 'peak_pos_range'})
)

song_range = song_range.merge(df[['title', 'performer', 'decade']], on=['title', 'performer'], how='left').drop_duplicates(['title', 'performer'])

range_per_decade = song_range.groupby('decade')['peak_pos_range'].mean()

range_per_decade.plot(kind="bar", figsize=(8,5))
plt.ylabel("Mean range of peak position")
plt.title("Peak position range per decade")
plt.show()
```
